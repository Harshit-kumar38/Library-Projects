```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
print('Setup is complete')
```

```python
# First make the 50 students data
np.random.seed(42) # I use this for the same data every time

num_students = 50

maths = np.random.randint(35,100,num_students)
english = np.random.randint(35,100,num_students)
science = np.random.randint(35,100,num_students)
sst = np.random.randint(35,100,num_students)
hindi = np.random.randint(35,100,num_students)

print('Maths marks: ',maths)
print('No of students: ',len(maths))
```

```python
# Fifty students names
student_names = [
    "Aarav Sharma", "Vivaan Gupta", "Aditya Singh", "Vihaan Kumar", "Arjun Verma",
    "Sai Patel", "Reyansh Mehta", "Ayaan Joshi", "Krishna Nair", "Ishaan Rao",
    "Ananya Iyer", "Diya Chopra", "Priya Mishra", "Riya Shah", "Anvi Pandey",
    "Saanvi Dubey", "Myra Tiwari", "Aarohi Jain", "Kavya Agarwal", "Pooja Reddy",
    "Rohit Yadav", "Mohit Sharma", "Amit Singh", "Sumit Kumar", "Ravi Verma",
    "Ajay Patel", "Vijay Mehta", "Sanjay Joshi", "Rakesh Nair", "Mahesh Rao",
    "Neha Iyer", "Sneha Chopra", "Rekha Mishra", "Geeta Shah", "Sunita Pandey",
    "Kavita Dubey", "Meena Tiwari", "Seema Jain", "Radha Agarwal", "Lata Reddy",
    "Harsh Yadav", "Kunal Sharma", "Nikhil Singh", "Rahul Kumar", "Gaurav Verma",
    "Shivam Patel", "Pranav Mehta", "Dhruv Joshi", "Yash Nair", "Dev Rao"
]

print('Total: ',len(student_names))
print('First: ',student_names[0])
print('Last: ',student_names[-1])
```

```python
# Now we make the dataframe

df = pd.DataFrame({
    'Student name': student_names,
    'Maths': maths,
    'English' : english,
    'Hindi' : hindi,
    'SST' : sst,
    'Science' : science
})

df.head()
```

```python
df['Total'] = df['Maths']+df['English']+df['SST']+df['Science']+df['Hindi']
df.head()

df['Percentage'] = (df['Total']/500)*100
df.head()
```

```python
def grade(per):
    if per>90:
        return 'A+'
    elif per>80:
        return 'A'
    elif per>70:
        return 'B'
    elif per>60:
        return 'C'
    elif per>50:
        return 'D'
    elif per>36:
        return 'E'
    else:
        return 'Fail'
```

```python
df['Grade'] = df['Percentage'].apply(grade)
df.head()
```

```python
df['Rank'] = df['Total'].rank(ascending=False,method='min').astype(int)
df.head()
```

```python
topper = df.loc[df["Total"].idxmax()]

print("🏆 School Topper")
print(topper)
```

```python
# Our database overview
print('Total Students: ',len(student_names))
print('Columns:',df.columns.to_list())
df.describe()
```

```python
# Subject Average
subject_avg = df[['Maths','English','Hindi','SST','Science']].mean()
subject_avg
```

```python
best_subject = subject_avg.idxmax()
print('Best Performing Subject: ',best_subject)
```

```python
worst = subject_avg.idxmin()
print('Worst Performing Subject: ',worst)
```

```python
# Top 10 Students
topper_list = df.sort_values('Rank')
topper_list.head(10)
```

```python
# Pass/Fail analysis
df['Result'] = np.where((df[['Maths','English','SST','Hindi','Science']]>36).any(axis=1),'Pass','Fail')

df['Result'].value_counts()
```

```python
# Pass Percentage
pass_percentage = ((df['Result']=='Pass').mean())*100
print('Pass percentage:',pass_percentage)
```

```python
# Subject wise bar chart
subject_avg = df[
    ["Maths", "Science", "English", "Hindi", "SST"]
].mean()

plt.figure(figsize=(8,5))
plt.bar(subject_avg.index, subject_avg.values)

plt.title("Subject-wise Average Marks")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")

plt.show()
```
